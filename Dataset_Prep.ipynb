```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import numpy as np
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
Master_df = pd.read_csv('./Data/TestDF20242025Q1 1.csv')
KPI_df = pd.read_csv('./Data/results_dfByMonth 1.csv')
```

```python
Master_df
```

```python
#filter for plant 2100
Master_df = Master_df[Master_df['Plant'] == 2100]

# filter for Dept STP
Master_df = Master_df[Master_df['Dept'] == 'STP']

Master_df
```

```python
# Drop the all rows with only 1 unique value
Master_df = Master_df.loc[:, Master_df.nunique() > 1]

#drop all columns with nan value
Master_df = Master_df.dropna(axis=1, how='all')

Master_df
```

```python
Master_df.MachineId.unique()
```

```python
Master_df = Master_df[["MachineId","MaterialCode","MaterialDesc","DimensionDesc","LSL","USL","MeasValue","CharId","Target","MeasDate","MeasYear","CheckBy"]]
```

```python
# Group by CharId and MachineId
results = []
RecordsWith1RecordOnlyMonthly = []
count = 0
for char_id, group in Master_df.groupby('CharId'):
    count+=1
    for machine_id, machine_group in group.groupby('MachineId'):
        if len(machine_group) >= 60:  # Only process if we have enough data points
            # Calculate mean and standard deviation
            mean = machine_group['MeasValue'].mean()
            std_dev = machine_group['MeasValue'].std(ddof=1)  # Sample std deviation

            if len(machine_group['MeasValue']) > 1:  # Ensure there are at least 2 data points
                deviations = [(x - mean) ** 2 for x in machine_group['MeasValue']]  # Squared deviations
                variance = sum(deviations) / (len(machine_group['MeasValue']) - 1)  # Sample variance
                std_dev_manual = variance ** 0.5  # Standard deviation
            else:
                RecordsWith1RecordOnlyMonthly.append({
                    'CharId': char_id,
                    'Time': machine_group['MeasYear'].iloc[0],  # Use first date in the group
                    'MachineId': machine_id,
                })
                continue  # Skip to next iteration if insufficient data

            # Specification limits
            LSL = machine_group['LSL'].iloc[0]
            USL = machine_group['USL'].iloc[0]
            Category = ''
            
            # Calculate Cp and Cpk
            cp = (USL - LSL) / (6 * std_dev) if std_dev > 0 else np.nan  # Handle case where std_dev is 0
            cpk = min((mean - LSL) / (3 * std_dev), (USL - mean) / (3 * std_dev)) if std_dev > 0 else np.nan
            
            if cp >= 1.0:
                Category = 'A'
            elif cp < 1.0:
                Category = 'B'
                
            # Append results
            results.append({
                'CharId': char_id,
                'Cp': cp,
                'Cpk': cpk,
                'Time': machine_group['MeasYear'].iloc[-1],  # Use last date in the group
                'MaterialCode': machine_group['MaterialCode'].iloc[-1],
                'MaterialDesc': machine_group['MaterialDesc'].iloc[-1],
                'DimensionDesc': machine_group['DimensionDesc'].iloc[-1],
                'MachineId': machine_id,
                'Category': Category,
                'std_dev': std_dev,
                'std_dev_manual': std_dev_manual,
                'mean': mean,
                'LSL': LSL,
                'USL': USL,
                'NumberOfData': len(machine_group['MeasValue'])
            })

# Convert results to dataframe
results_df = pd.DataFrame(results)

# Display the results
print(f"Generated {len(results_df)} records from {len(Master_df.groupby(['CharId', 'MachineId']))} CharId-MachineId combinations")
results_df.head()
```

```python
# can you groupby char id and machine id and append the cp, cpk, std_dev, std_dev_manual and mean values from the results_df to Master_df
Master_df = Master_df.merge(results_df[['CharId', 'MachineId', 'Cp', 'Cpk', 'std_dev', 'std_dev_manual', 'mean']], on=['MachineId','CharId' ], how='left')
Master_df = Master_df.drop_duplicates(subset=['MachineId', 'CharId', 'MeasDate'], keep='last')
```

```python
# Group by MaterialDesc and CharId
df = Master_df.copy()
results = []

for Machine in df['MachineId'].unique():
    for material in df['MaterialDesc'].unique():
        for char_id in df[(df['MaterialDesc'] == material) & (df['MachineId'] == Machine)]['CharId'].unique():
            # Filter the DataFrame for the current Material and CharId
            Filter = df[(df['MaterialDesc'] == material) & (df['CharId'] == char_id)]
            
            if len(Filter) < 30:  # Ensure there are enough data points for analysis
                print(f"Not enough data for Machine:{Machine} Material: {material}, CharId: {char_id}")
                continue
            
            # Fit the ANOVA model
            model = ols('MeasValue ~ C(CheckBy)', data=Filter).fit()
            anova_results = sm.stats.anova_lm(model)
            
            # Extract variance components
            repeatability_mean_sq = anova_results['mean_sq'][0]  # Mean square for CheckBy
            repeatability_st_dv = math.sqrt(repeatability_mean_sq)
            
            # Calculate Tolerance
            tolerance = Filter['USL'].unique()[0] - Filter['LSL'].unique()[0]
            
            # Calculate Gage R&R
            gage_rr = round((repeatability_st_dv / tolerance) * 100, 2)
            
            # Store results
            results.append({
                'MachineId': Machine,
                'MaterialDesc': material,
                'CharId': char_id,
                'Gage_RR': gage_rr,
                'Tolerance':tolerance,
                
            })

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Display the results
print(results_df)
```
